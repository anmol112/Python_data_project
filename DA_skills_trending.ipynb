```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
from datasets import load_dataset
import ast
import matplotlib.ticker as ticker
import seaborn as sns

```

```python
ds = pd.read_pickle('datasets/ds_cleaned_2')
```

```python
DA = ds[ds['job_title_short'] == 'Data Analyst']
```

```python
DA['total jobs'] = DA['job_title_short'].count()
```

```python
DA
```

```python
DA_exploded = DA.explode('job_skills_copy')
```

```python
DA_exploded
```

```python
DA_exploded.drop(columns='job_type_skills_copy' , inplace= True)
```

```python
DA_exploded['month'] = DA_exploded['job_posted_date'].dt.month
```

```python
def y_n(string):
    if string == 'python':
        return 'y'
    else : 
        return 'n'
```

```python
top_10_skills_dataframe = DA_exploded.groupby('job_skills_copy' ).agg(

    counts = ('job_skills_copy' , 'count')
)
```

```python
DA['total jobs'].unique()
```

```python
top_10_skills_dataframe
```

```python
top_10_skills_dataframe = top_10_skills_dataframe.reset_index()
```

```python
top_10_skills_dataframe = top_10_skills_dataframe.sort_values('counts' , ascending=False).head(10)
```

```python
top_10_skills_list = top_10_skills_dataframe['job_skills_copy'].to_list()
```

```python
top_10_skills_list
```

```python
prac = DA_exploded.groupby(['month' , 'job_skills_copy']).agg(
    count_postings = ('job_skills_copy' , 'count')
).reset_index()
```

```python
for_percent = DA
```

```python
for_percent['month'] = for_percent['job_posted_date'].dt.month
```

```python
jobs_post_monthly = for_percent.groupby('month').agg(
    jobs_per_month = ('job_title_short' , 'count')
)
```

```python
jobs_post_monthly
```

```python
my_list = top_10_skills_list
unique_list = []
for item in my_list:
    if item not in unique_list:
        unique_list.append(item)


unique_list
```

```python
merged_DA = None
for val in unique_list :
    line = val
    pivot_fnc = prac[prac['job_skills_copy'] == line].copy()
    pivot_fnc.rename(columns={'count_postings': line} , inplace= True)
    pivot_fnc.drop(columns='job_skills_copy' , inplace=True)

    if merged_DA is None :
        merged_DA = pivot_fnc.copy()
    else :
        merged_DA = pd.merge(merged_DA , pivot_fnc , on= 'month').copy()

merged_DA
```

```python
DA_skills_trend_dataset = pd.merge(merged_DA , jobs_post_monthly , on='month'  )
```

```python
DA_skills_trend_dataset
```

```python
DA_skills_trend_dataset['sql'] = (DA_skills_trend_dataset['sql'] / DA_skills_trend_dataset['jobs_per_month']) * 100
DA_skills_trend_dataset['excel'] = (DA_skills_trend_dataset['excel'] / DA_skills_trend_dataset['jobs_per_month']) * 100
DA_skills_trend_dataset['python'] = (DA_skills_trend_dataset['python'] / DA_skills_trend_dataset['jobs_per_month']) * 100
DA_skills_trend_dataset['tableau'] = (DA_skills_trend_dataset['tableau'] / DA_skills_trend_dataset['jobs_per_month']) * 100
DA_skills_trend_dataset['power bi'] = (DA_skills_trend_dataset['power bi'] / DA_skills_trend_dataset['jobs_per_month']) * 100
DA_skills_trend_dataset['r'] = (DA_skills_trend_dataset['r'] / DA_skills_trend_dataset['jobs_per_month']) * 100
DA_skills_trend_dataset['sas'] = (DA_skills_trend_dataset['sas'] / DA_skills_trend_dataset['jobs_per_month']) * 100
DA_skills_trend_dataset['powerpoint'] = (DA_skills_trend_dataset['powerpoint'] / DA_skills_trend_dataset['jobs_per_month'])  * 100
DA_skills_trend_dataset['word'] = (DA_skills_trend_dataset['word'] / DA_skills_trend_dataset['jobs_per_month'])  * 100
DA_skills_trend_dataset['sap'] = (DA_skills_trend_dataset['sap'] / DA_skills_trend_dataset['jobs_per_month']) * 100
```

```python
DA_skills_trend_dataset
```

```python
DA_skills_trend_dataset.to_pickle('data_graphs/DA_skills_trending')
```

```python
merged_DA_2 =  DA_skills_trend_dataset.loc[: , [  'sql' , 'excel' , 'python' , 'tableau' , 'power bi' , 'r' , 'sas']]
```

```python
merged_DA_2
```

```python

merged_DA_2.plot(kind='line' , figsize=(8 ,5))
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y , pos : f'{int(y)}%'))
ax.set_xlim(0 ,12)
plt.title('Trending top skills for Data Analyst')
plt.xlabel('Months')
plt.ylabel('Likelihood of skill requested in each posting')
ax.set_xticks(range(0, 13)) 
sns.despine()
```
