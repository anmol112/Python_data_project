```python
def lstr_list(string):
    line = ""
    final_list = []
    switch = 'off'

    if string is None or string == "" :
        return []

    else:
        for val in string:
            if val == "'" and switch == 'off':
                switch = 'on'  # Corrected assignment operator

            elif val == "'" and switch == 'on':
                switch = 'off'
                final_list.append(line)
                line = ""

            elif val != "'" and switch == 'on':
                line += val  # Corrected to concatenate characters



    return final_list
        
```

```python
# Function to split strings based on single quotes
def lstr_list(string):
    if string is None:
        return []  # Return an empty list if the input is None
    
    line = ""
    final_list = []
    switch = 'off'

    for val in string:
        if val == "'" and switch == 'off':
            switch = 'on'
        elif val == "'" and switch == 'on':
            switch = 'off'
            if line:
                final_list.append(line)
                line = ""
        elif val != "'" and switch == 'on':
            line += val

    if line:
        final_list.append(line)

    return final_list

# Apply function to each cell in the 'col' column
ds_copy['job_skills_copy'] = ds_copy['job_skills'].apply(lstr_list)
type(ds_copy['job_skills_copy'][1])

```
