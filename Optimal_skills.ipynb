```python
optimal_skills_2 = optimal_skills.groupby('job_skills_copy').agg(
    count = ('job_skills_copy' , 'count'),
    salary = ('salary_year_avg' , 'median')
)
```

```python
optimal_skills_2.reset_index(inplace=True)
```

```python
k = ds_copy['job_type_skills_copy']
```

```python
k_list = k.to_list()
```

```python
k_list[20]
```

make a function for dict to combine dicts 

```python
#code that join items of all dict into one unique dict
final_list = {}

for one in k_list:
    # can make dict combined one to final 
    for k in one :
        if k in final_list.keys():
            #keyname == val
            #checking if key belongs in final keys

            for value in one[k]:

                if value in final_list[k]:
                    pass
                    #because same value h final m 

                else :
                    #update krde value nhi h is final[keys] me 
                    final_list[k].append(value)

        else:
            #update krde direct key value pair 

            final_list[k] = one[k]


```

```python
final_list
```

```python
#-----------------------------------
#convert dict into desired dataframe 
data = [(key, value) for key, values in final_list.items() for value in values]

# Convert the list of tuples into a DataFrame
skill_types = pd.DataFrame(data, columns=["Key", "Value"])

```

```python
skill_types.drop_duplicates(inplace=True)
```

```python
optimal_skills_2.drop_duplicates(inplace=True)
```

```python
optimal_skills_2.sort_values(by='count' , ascending=False)
```

```python
skill_types
```

```python
optimal_skills_final = pd.merge(optimal_skills_2 , skill_types , left_on='job_skills_copy',right_on='Value')
```

```python
optimal_skills_final['count'] = optimal_skills_final['count'] * 100/total_jobs
```

```python
optimal_skills_final_2 = optimal_skills_final.sort_values(by='count', ascending=False).head(15)
```

```python
optimal_skills_final_2.to_pickle('data_graphs/optimal_skills')
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(optimal_skills_final_2 , x='count', y='salary' , hue='Key')
plt.xlabel('Likelihood of Skill Occurring in Each Job Posting (%)')
plt.ylabel('Average Salary')
plt.title('Best Computer Science Skills ')

texts = []
for i ,text in enumerate(optimal_skills_final_2['job_skills_copy']):
    texts.append(plt.text(optimal_skills_final_2['count'].iloc[i] , optimal_skills_final_2['salary'].iloc[i] , text))

adjust_text(texts , arrowprops=dict(arrowstyle="->" , color='black' , lw=1))

ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x , pos:f'{x}%'))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y , pos:f'{int(y/1000)}K'))
sns.despine()
plt.legend(title = None)
```
